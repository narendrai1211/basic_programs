```python
import pandas as pd
```

```python
df = pd.read_excel('/content/Share Market.xlsx')
```

```python
filtered_df = df[['Must have companies', 'Price', 'Quantity', 'Total Price']]
```

```python
filtered_df
limit_price = 10000
stock_qty_limit = []
stock_total_price_list = []

for i,row in filtered_df.iterrows():
  stock_price = row['Price']
  stock_name = row['Must have companies']
  stock_per_limit = int(limit_price / int(stock_price))
  stock_total_cost = stock_per_limit * stock_price
  final_total_price = stock_per_limit * stock_price  

  while final_total_price > limit_price:
    stock_per_limit -= 1
    final_total_price = stock_per_limit * stock_price  

  stock_qty_limit.append(int(stock_per_limit))
  stock_total_price_list.append(int(final_total_price))


filtered_df['Quantity'] = stock_qty_limit
filtered_df['Total Price'] = stock_total_price_list
```
