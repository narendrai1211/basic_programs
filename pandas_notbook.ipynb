```python
# SELECT ALL THE RIDES WHICH STARTED FROM NEW YORK
df_ = df.loc[
    (df['MILES*'] > 10) & (df['START*'].isin(['New York', 'Cary'])),
    ['START*', 'MILES*']] 
df_.reset_index(drop=True, inplace=True) # to drop the index column which is created in prev df 
df_
```

```python
df_short = df['MILES*'][df['MILES*'] < 5]
df_long = df['MILES*'][df['MILES*'] >= 5]
df_short.to_list()
df_long.to_list()
df['Short_trip']= df_short
df['Long_trip'] = df_long
```

```python
mean_with_agg = df.groupby('START*')['MILES*'].agg('mean')
mean_simplified = df.groupby('START*')['MILES*'].mean()
```
